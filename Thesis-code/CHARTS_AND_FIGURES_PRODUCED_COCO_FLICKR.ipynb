```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from your table
data = {
    "Model": ["BLIP", "GPT-2", "PIX2STRUCT", "ANN Model+Weighted Summarization"],
    "BLEU": [0.305, 0.317, 0.054, 0.322],
    "METEOR": [0.302, 0.327, 0.165, 0.328],
    "ROUGE-1 f": [0.401, 0.404, 0.227, 0.452],
    "ROUGE-2 f": [0.163, 0.158, 0.029, 0.187],
    "ROUGE-l f": [0.369, 0.369, 0.206, 0.415]
}

# Creating DataFrame
df = pd.DataFrame(data)

reordered = ["PIX2STRUCT", "BLIP", "GPT-2", "ANN Model+Weighted Summarization"]
df['Model'] = pd.Categorical(df['Model'], categories=reordered, ordered=True)
df.sort_values('Model', inplace=True)

# Transpose the DataFrame to group by metrics
df.set_index('Model', inplace=True)
df_t = df.transpose()  # Transposing the DataFrame

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
bars = df_t.plot(kind='bar', ax=ax, width=0.8)
ax.set_title('Performance Comparison of Models Across Metrics on COCO Dataset')
ax.set_ylabel('Scores')
ax.set_xlabel('Metrics')
plt.xticks(rotation=0)
plt.legend(title='Models',loc='upper left')

# Annotating each bar with its value
for container in bars.containers:
    labels = [f'{v.get_height():.3f}' for v in container]
    # Apply bold formatting if this is the ANN Model+Weighted Summarization
    if container.get_label() == 'ANN Model+Weighted Summarization':
        ax.bar_label(container, labels=labels, label_type='edge', padding=7, weight='bold')
    else:
        ax.bar_label(container, labels=labels, label_type='edge', padding=1)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Epoch data
epochs = list(range(1, 21))

# Loss and accuracy values
loss = [
    0.7552, 0.5358, 0.4849, 0.3880, 0.3393, 0.2890, 0.2754,
    0.2052, 0.1641, 0.1287, 0.1761, 0.1673, 0.1470, 0.1031,
    0.0447, 0.0682, 0.0428, 0.0171, 0.0148, 0.0276
]

accuracy = [
    0.5665, 0.7099, 0.7580, 0.8268, 0.8498, 0.8819, 0.8635,
    0.9232, 0.9392, 0.9587, 0.9312, 0.9381, 0.9438, 0.9702,
    0.9839, 0.9817, 0.9885, 0.9989, 0.9966, 0.9931
]

# Create the plots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot loss
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color='tab:red')
ax1.plot(epochs, loss, 'o-', color='tab:red', label='Loss')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create twin axis for accuracy
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', color='tab:blue')
ax2.plot(epochs, accuracy, 's-', color='tab:blue', label='Accuracy')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title and legend
plt.title('COCO: Training Loss and Accuracy per Epoch')
fig.tight_layout()

plt.show()
```
